7. Escreva um programa que calcula o fatorial de um dado número N.

```python
num = int(input("Insira um número: "))

if num == 0:
  print("0! = 1")

elif num == 1:
  print("1! = 1")

elif num > 1:
  fatorial = num
  for i in range(num,1,-1):
    fatorial = fatorial*i

  print(f"{num}! = {fatorial}")
```

16. Faça um programa que recebe o número real x como entrada e devolva uma
aproximação do arco tangente de x (em radianos) através da série:

arctan(x) = x-(x^3/3)+(x^5/5)-(x^7/7)+....

Considere a aproximação para 50 termos.

```python
x = float(input("X: "))

i = 0
divisor = 3
arctan = x

while i < 50:
  if i%2 == 0:
    arctan-=(x**divisor)/divisor

  else:
    arctan+=(x**divisor)/divisor
  i+=1
  divisor+=2

print(f"arctan({x}) =", arctan)
```

21. Faça um programa que peça um número inteiro e determine se ele é ou não
um número primo. Um número primo é aquele que é divisível somente por ele
mesmo e por 1.

```python
num = int(input("Insira um número: "))
contador = 0

for i in range(1,num+1):
  if num%i == 0:
    contador+=1

if contador == 2 and num != 0:
  print(num,"é primo")

else:
  print(num,"não é primo")
```
